```python
#Import pathlib and csv libraries
from pathlib import Path
import csv

#Set the file path
csvpath=Path("Resources/budget_data.csv")

#Initialize list of records
records=[]

#Open csv file as an object (use "read" mode)
with open(csvpath, 'r') as csvfile:
    
    #Use csv.reader() fucntion to read and return csvreader object
    csvreader=csv.reader(csvfile, delimiter=",")
    
    # Iterate through the data and search for the number the user inputted. Remember to skip the header of the CSV.
    csv_header = next(csvreader)
    
    #Initiate metrics
    month_count=0                    #Total number of months
    total=0                          #Net total amount of Profit/Losses over the entire period
    
    change=0                         #Changes in Profit/Losses
    previous_month=0                 #Value of Profit/Losses in previous month
    total_change=0                   #Total of changes over the entire period
    average_change=0                 #Average of the changes in Profit/Losses over the entire period
    
    greatest_increase_profits=0      #The greatest increase in profits over the entire period
    greatest_incr_date=0             #Month and year of the greatest increase in profit
    greatest_decrease_losses=0       #The greatest decrease in losses over the entire period
    greatest_decr_date=0             #Month and year of the greatest decrease in loss
    
    #Using for loops to run through the data and compute required metrics
    for row in csvreader:
        month_count+=1
        total += int(row[1])
        
        #Change in P/L current month
        if previous_month==0:
            previous_month=int(row[1])
            
        change=int(row[1]) - previous_month
        total_change += change
        previous_month = int(row[1])
            
            #Find the greatest increase in profits
        if change>0:
            if greatest_increase_profits==0:
                greatest_increase_profits=change
                greatest_incr_date=row[0]
            elif change>greatest_increase_profits:
                greatest_increase_profits=change
                greatest_incr_date=row[0]
                    
            #Find the greatest decrease in losses
        elif change<0:
            if greatest_decrease_losses==0:
                greatest_decrease_losses=change
                greatest_decr_date=row[0]
            elif change<greatest_decrease_losses:
                greatest_decrease_losses=change
                greatest_decr_date=row[0]
                    
    average_change=round(total_change/(month_count-1),2)
    
#Print the analysis to the terminal
print('Financial Analysis')
print('--------------------------')
print(f'Total Months: {month_count}')
print(f'Total: ${total}')
print(f'Average  Change: ${average_change}')
print(f'Greatest Increase in Profits: {greatest_incr_date} (${greatest_increase_profits})')
print(f'Greatest Decrease in Profits: {greatest_decr_date} (${greatest_decrease_losses})')
```

```python
#Export a text file with the results
output_path=Path("Outputresult.txt")

with open(output_path, 'w') as output:
    output.write('Financial Analysis\n')
    output.write('--------------------------\n')
    output.write(f'Total Months: {month_count}\n')
    output.write(f'Total: ${total}\n')
    output.write(f'Average  Change: ${average_change}\n')
    output.write(f'Greatest Increase in Profits: {greatest_incr_date} (${greatest_increase_profits})\n')
    output.write(f'Greatest Decrease in Profits: {greatest_decr_date} (${greatest_decrease_losses})\n')   
```
